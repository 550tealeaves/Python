# Flow Control Statements
## if statement

- if statement = most common flow control statement
- if statement clause (block after "if") executes condition if True, but skips the clause if False
- Contains
- - if
  - condition that is either True or False
  - colon
  - if block (indented block of code)

```python
name = 'Luceryn'
if name == 'Luceryn':
    print(f"Hi, {name}")
    print('Good morning, Luceryn')
```

## Else
- Runs only when if clause is False
- Doesn't have a condition
- Contains
- - "else"
  - colon
  - indented block (else clause/block) on next line

```python
name = 'Arya'
if name == 'Lyanna':
    print(f"Hi, {name}")
else:
    print("Hi, stranger") # prints this statement b/c if clause is False
```

## Elif 
- Use if you want one of many possible clauses to run
- follows an if or elif statement
- checked only if all previous conditions are False
- Contains
- - "elif"
  - condition that's True/False
  - colon
  - elif clause/block
- no guarantee that at least 1 clause runs
- Only 1 or 0 clauses run in a chain of elif statements

```python
name = 'Visenya'
age = 33
if name == 'Visenya':
    print('Hi, Visenya.')
elif age < 12:
    print('You are not Visenya, kiddo.')
```

```python
# Introducing an error
name = 'Carol'
age = 3000
if name == 'Alice':
    print('Hi, Alice.')
elif age < 12:
    print('You are not Alice, kiddo.')
elif age > 2000: # prints out this statement b/c order of elif matters & this is the first one that is True
    print("Unlike you, Alice is not an undead immortal")
elif age > 100:
    print("you are not Alice, baba")
```

```python
name = 'Carol'
age = 3000
if name == 'Alice':
    print('Hi, Alice.')
elif age < 12:
    print('You are not Alice, kiddo.')
else:
    print('You are neither Alice nor a little kid.')
```
