## Variables 
- Can store a single value of an evaluated expression to be used later
- Like a labeled box w/ a value inside



```python
spam = 40
spam
#40
eggs = 2
spam + eggs #40 + 2
#42
spam + eggs + spam #40 + 2 + 40
#82
spam = spam + 2 #40 + 2
spam
#42
```

- Variable initialized (created) 1st time a value is stored w/in
- Can then be used w/ other variables and values
- If you assign new value to a variable = will forget the old value  overwriting




```python
spam = 'Hello, from the other siiiide'
spam
```

## Print() function
- Displays value inside the ()
- Python calls the print() function & a value is passed to the function
- - Argument = value passed into a function
- Won't print the quotes from a string b/c quotes only demarcate start/end pts of string
- Can do print() to display a blank line

## Input() function
- Asks user to enter in text and press enter
- Evaluates string identical to user's text & assigns that value to a variable (in the above example)
- '>' passed into function means that > prompt appears = indicates user to type something
- Think of input() function as expression that evaluates to whatever string the user typed
